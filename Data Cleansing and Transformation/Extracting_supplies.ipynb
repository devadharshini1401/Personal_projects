```python
import pandas as pd
```

```python
read_file = pd.read_excel("/Users/deva/Desktop/Analysis/Hosp+pract/market.xlsx")
```

```python
a = read_file['supplieslist'].dropna()
```

```python
df_supply = pd.DataFrame(a.str.split('|').tolist()).stack()
df_supply
```

```python
df = set()

for i in df_supply:
    df.add(i)
df
```

```python
df_list = list(df)
len(df_list)
```

```python
count = 1
df_new = []
for i in df_list:
    if i not in df_new:
        df_new.append([count,i])
        count+=1
df_new
```

```python
supply = pd.DataFrame(df_new,columns = ['Equip_id','Equip_desc'])
```

```python
output = []
for desc in supply['Equip_desc']:
    if isinstance(desc, str):
        words = desc.split()
        code = ''
        for word in words:
            for char in word:
                if char.isalpha():
                    code += char.upper()  # Take the alphabetic characters only
                    break  # Move to the next word after finding the first alphabetic character
        output.append(code[:3])  # Limit to the first 3 characters
    else:
        output.append('')

# Add the new column to the supply dataframe, not df
supply['Equip_code_new'] = output
```

```python
supply
```

```python
supply['Equip_code'] = supply['Equip_code_new'] + supply['Equip_id'].astype(str)
```

```python
supply.drop(columns = ['Equip_code_new'],inplace = True)
```

```python
supply.to_csv("Supplies.csv",index = False)
```
