```python
from sqlalchemy import create_engine,Table,Column,String,Integer,Date,MetaData
```

```python
metadata = MetaData()
```

```python
engine = create_engine('sqlite:///db_deva.db')
```

```python
hospital_table = Table(
    'hospital1', metadata,
    Column('Hospital_id', Integer, primary_key=True),
    Column('Hospital_name', String),
    Column('Practice_name',String),
    Column('Create_date', Date)
)
```

```python
con = engine.connect()
```

```python
metadata.create_all(engine)
```

```python
read_file1 = pd.read_excel('market.xlsx')
```

```python
df = read_file1[['businessname','practicename','practiceaddress1','practiceaddress2','practicecity','practicestate','practicezip9code','telephonenumber']]
```

```python
df
```

```python
sql_query = "SELECT * FROM hospital1"
hosp = pd.read_sql(sql_query, engine)
```

```python
hosp
```

```python
# Assuming 'df' and 'hosp' are DataFrames and you want to perform an inner join
joined_df = pd.merge(df, hosp, left_on=['businessname','practicename'], right_on=['Hospital_name','Practice_name'], how='inner')
```

```python
joined_df
```

```python
joined_df['branch_id'] = range(1500,len(joined_df)+1500)
```

```python
joined_df.rename(columns={'practiceaddress1':'Address',
                          'practiceaddress2':'Address1','practicecity':'City','practicestate':'State',
                          'practicezip9code':'Zipcode','telephonenumber':'Phone','branch_id':'Branch_id'
                         
                         },inplace = True)
```

```python
joined_df
```

```python
new_df = joined_df[['Branch_id','Hospital_id','Address','Address1','City','State','Zipcode','Phone']]
```

```python
new_df
```

```python
new_df.to_sql('Branch1', con=engine, if_exists='replace', index=False)
```
